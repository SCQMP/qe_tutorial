```python
# %%
import numpy as np
import matplotlib.pyplot as plt
from qwt.qe import read_band_structure

kpath, kpoints, enk = read_band_structure("./bands/Si.xml")
ymin, ymax = -3, 3 # eV

nkl = 30
nnkl = 5
kticks = kpath[nkl*np.arange(nnkl)]
# K_POINTS {crystal_b}
# 5
#   0.0000 0.5000 0.0000 30  !L
#   0.0000 0.0000 0.0000 30  !G
#   -0.500 0.0000 -0.500 30  !X
#   -0.375 0.2500 -0.375 30  !U
#   0.0000 0.0000 0.0000 30  !G
klabels = ['$L$', '$\Gamma$', '$X$', '$U$', '$\Gamma$']
```

```python



plt.figure(figsize=(4,4),dpi=300,layout='constrained')
plt.axhline(0, color='lightgray', linestyle='--', lw=0.5)
plt.grid(axis='x', color='lightgray', linestyle='--', lw=0.5)
for ib in range(enk.shape[1]):
    plt.plot(kpath, enk[:, ib], "k-")
plt.xticks(kticks, klabels)
plt.xlim(kpath[0], kpath[-1])
plt.ylim(ymin, ymax)
plt.ylabel("Energy (eV)")
plt.savefig("bands.png")

# %%
```
